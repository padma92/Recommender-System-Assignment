```python
#UBCF
import pandas as pd
import numpy as np
```

```python
# making data frame from csv file
book_df = pd.read_csv("book (1).csv")
 
# sorting by first name
book_df.sort_values("UserId", inplace = True)
 
# dropping ALL duplicate values
book_df.drop_duplicates(subset ="UserId",
                     keep = False, inplace = True)
 
# displaying data
book_df[0:10]
```

```python
#number of unique users in the dataset
len(book_df.UserId.unique())
```

```python
len(book_df.BookTitle.unique())
```

```python
user_book_df = book_df.pivot(index='UserId',
                        columns='BookTitle',
                        values='BookRating').reset_index(drop=True)
```

```python
user_book_df
```

```python
#Impute those NaNs with 0 values
user_book_df.fillna(0, inplace=True)
```

```python
user_book_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_book_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = book_df.UserId.unique()
user_sim_df.columns = book_df.UserId.unique()
```

```python
user_sim_df.iloc[0:10, 0:10]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:10, 0:10]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:10]
```

```python
#IBCF
#Fetching ratings for Toy Story
Murder_Book_user_rating = user_book_df['Apocalypse Wow']  
```

```python
Murder_Book_user_rating
```

```python
#Finding the correlation with different movies
similar_to_Murder_Book = user_book_df.corrwith(Murder_Book_user_rating) 
```

```python
similar_to_Murder_Book.head(10) 
```

```python
corr_Murder_Book = pd.DataFrame(similar_to_Murder_Book, columns=['Correlation'])
corr_Murder_Book.dropna(inplace=True)
corr_Murder_Book.head()
```

```python
corr_Murder_Book[corr_Murder_Book['Correlation'] > 0].sort_values(by='Correlation', ascending=False).head(10)  
```
